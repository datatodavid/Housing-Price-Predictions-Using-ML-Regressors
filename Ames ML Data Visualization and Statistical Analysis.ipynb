```python
def num_analyses(df):
    num_analysis = pd.DataFrame([], columns=['min_','mean_','median_','max_','sd_','mode_','mode_freq','sec_mode', 'sec_mode_freq','num_unique','zeroes', 'NAs_'])
    df = df.select_dtypes(include = np.number)
    for col_ in list(df.columns):
        num_analysis.loc[col_] = [min(df[col_]), round(np.mean(df[col_]),1), np.nanmedian(df[col_]), max(df[col_]),\
                                  np.std(df[col_]), df[col_].mode().values[0], sum(df[col_]==df[col_].mode().values[0]),\
                                  df[col_][df[col_]!=df[col_].mode().values[0]].mode().values[0],\
                                  sum(df[col_]==df[col_][df[col_]!=df[col_].mode().values[0]].mode().values[0]),\
                                  len(df[col_].unique()), sum(df[col_]==0), sum(df[col_].isna())]
    return num_analysis

ames_num_stats = num_analyses(ames_dum_filter_full)    
        
ames_num_NAs = ames_num_stats[ames_num_stats['NAs_']!=0].sort_values(['NAs_'], ascending=False)
ames_num_NAs
ames_num_zeroes = ames_num_stats[ames_num_stats['zeroes']!=0].sort_values(['zeroes'], ascending=False)
ames_num_zeroes

ames_num_stats = round(ames_num_stats.sort_values(['zeroes'], ascending=False).T,2)
ames_num_stats
```

```python
def char_analyses(df):
    char_analysis = pd.DataFrame([], columns=['mode_','mode_freq','sec_mode','sec_mode_freq','num_unique','choices_','NAs_'])
    df = df.select_dtypes(exclude = np.number)
    for col_ in list(df.columns):
        char_analysis.loc[col_] = [df[col_].mode().values[0], sum(df[col_]==df[col_].mode().values[0]), \
                                  df[col_][df[col_]!=df[col_].mode().values[0]].mode().values[0],\
                                  sum(df[col_]==df[col_][df[col_]!=df[col_].mode().values[0]].mode().values[0]),\
                                  len(df[col_].unique()), df[col_].unique(), sum(df[col_].isna())]
    return char_analysis

ames_char_stats = char_analyses(ames_orig)
                    
ames_char_stats        

ames_char_NAs = ames_char_stats[ames_char_stats['NAs_']!=0].sort_values(['NAs_'], ascending=False)
ames_char_NAs
```

```python
import scipy.stats
# from sklearn.preprocessing import StandardScaler

dist_names = ['weibull_min','norm','weibull_max','beta',\
              'invgauss','uniform','gamma','expon',\
              'lognorm','pearson3','triang']

ames_num = ames_dum_filter_full.select_dtypes(include = np.number)
ames_num = ames_num.drop(ames_num_NAs.index, axis=1)

# scaler = StandardScaler()
# ames_scaled_num = pd.DataFrame(scaler.fit_transform(ames_num), columns=ames_num.columns, index=ames_num.index)
# ames_scaled_num

# 11 equi-distant bins of observed Data 
results123 = pd.DataFrame([], columns = ['Feature','Distribution1', 'chi_sq1',\
                                           'Distribution2', 'chi_sq2', \
                                          'Distribution3', 'chi_sq3'] )

# Loop through candidate distributions
for col_ in list(ames_num.columns):
    percentile_bins = np.linspace(0,100,11)
    chi_square_statistics = []
    percentile_cutoffs = np.percentile(ames_num[col_], percentile_bins)
    observed_frequency, bins = (np.histogram(ames_num[col_], bins=percentile_cutoffs))
    cum_observed_frequency = np.cumsum(observed_frequency)

    

    #     size = len(ames_scaled_num.loc[scipy.stats.zscore(ames_scaled_num[col_].abs()<=3)])
    try:
        for distribution in dist_names:
            # Set up distribution and get fitted distribution parameters
            dist = getattr(scipy.stats, distribution)
            param = dist.fit(ames_num[col_])
    #       print("{}\n{}\n".format(dist, param))


            # Get expected counts in percentile bins
            # cdf of fitted sistrinution across bins
            cdf_fitted = dist.cdf(percentile_cutoffs, *param)
            expected_frequency = []
            for bin in range(len(percentile_bins)-1):
                expected_cdf_area = cdf_fitted[bin+1] - cdf_fitted[bin]
                expected_frequency.append(expected_cdf_area)

            # Chi-square Statistics
            expected_frequency = np.array(expected_frequency) * len(ames_num)
            cum_expected_frequency = np.cumsum(expected_frequency)
            ss = sum (((cum_expected_frequency - cum_observed_frequency) ** 2) / cum_observed_frequency)
            chi_square_statistics.append(ss)

        #Sort by minimum chi-square statistics
        results = pd.DataFrame()
        
        results['Distribution'] = dist_names
        results['chi_square'] = chi_square_statistics
        results.sort_values(['chi_square'], inplace=True)
        results1 = pd.DataFrame([[ames_num[col_].name, results.iloc[0][0], results.iloc[0][1],\
                                  results.iloc[1][0], results.iloc[1][1], \
                                 results.iloc[2][0], results.iloc[2][1]]], \
                                columns = ['Feature','Distribution1', 'chi_sq1',\
                                           'Distribution2', 'chi_sq2', \
                                          'Distribution3', 'chi_sq3'] )
        results123 = results123.append(results1)
    except:
        results123 = results123.append(pd.DataFrame([[ames_num[col_].name, 'NaN', 'NaN', 'NaN', 'NaN', 'NaN', 'NaN']],\
                                                   columns = ['Feature','Distribution1', 'chi_sq1',\
                                           'Distribution2', 'chi_sq2', \
                                          'Distribution3', 'chi_sq3']))

# print ('\nDistributions listed by Betterment of fit:')
# print ('............................................')
# print (results)
# results
results123
# results2
```
